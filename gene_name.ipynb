```python
import re
# 存储匹配结果的集合，集合自动去重
matches_set = set()
pattern = re.compile(r';gene=[A-Za-z0-9]+;')

with open('D:\\hqh_sci_cailiao\\gene\\GCF_000001405.40_GRCh38.p14_genomic.gff', 'r') as file:
    i=0
    for line in file:
        # 使用正则表达式匹配gene=KIR2DS*;格式的字符串
        #match = re.search(r'gene=KIR2DS[^;]*;', line)
        matches=pattern.findall(line)
        for match in matches:
            # 提取匹配的部分并去掉gene=和;，然后添加到集合中
            gene_value = match.replace(';gene=', '').replace(';', '')
            #print(gene_value)
            i+=1
            if i%10000==0:
                print(i)
            matches_set.add(gene_value)

# 将集合转换为列表并保存到新的文件中
with open('D:\\hqh_sci_cailiao\\gene\\matched_genes_h2.txt', 'w') as output_file:
    for gene in matches_set:
        output_file.write(gene + '\n')
```
